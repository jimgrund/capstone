```python
import pandas as pd
university_data = pd.read_excel('Labeled Courses Orig.xlsx', header=0)
university_data
```

```python
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.externals import joblib

# Vectorizer
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(list(university_data['Course Description']))
terms = vectorizer.get_feature_names()

# Kmeans Model
model = KMeans(n_clusters=30, max_iter=20000, random_state=1) 
model.fit(X)
clusters = model.labels_.tolist()

# Model save
joblib.dump(model,  'desc_cluster.pkl')
```
